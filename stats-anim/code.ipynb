```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Generate some data
np.random.seed(0)
X = np.random.rand(100, 1)
y = 3 + 2 * X + np.random.randn(100, 1) / 1.5
X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance

# Normal equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

fig, ax = plt.subplots(dpi=300)

def update(frame):
    ax.clear()
    # Plotting the data
    ax.scatter(X, y, label='Training data')
    # Plotting the line
    X_new = np.array([[0], [1]])
    X_b_new = np.c_[np.ones((2, 1)), X_new]  # adds a column of 1s to X_new
    y_new = X_b_new.dot(theta_best)
    ax.plot(X_new, y_new, "r-", linewidth=2, label='Prediction')
    ax.legend(loc='upper right')
    return ax

animation = FuncAnimation(
    fig=fig,
    func=update,
    frames=range(20),
    interval=100  # in milliseconds
)
animation.save('convergence-LR.gif')
```
