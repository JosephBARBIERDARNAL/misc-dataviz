```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
df['workingPerWeek'] = df['workingPerWeek'].astype(float)
df.head()
```

```python
fig, ax = plt.subplots(dpi=300, figsize=(15,10))

ax.stem(df['country'], df['workingPerWeek'])
ax.set_ylim(0, 45)

plt.show()
```
